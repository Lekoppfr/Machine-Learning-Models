### the following is the code of K-Means Algorithms which be used to cluster different cases in prognosis or any other function on primlinary level.

```python
import matplotlib.pyplot as plt
import numpy as np
plt.jet()
import sklearn.datasets as datasets
x,y=datasets.make_blobs(centers=4,cluster_std=0.5,random_state=0)
plt.scatter(x[:,0],x[:,1])
```

trying to insert colors in the different clusters

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

training the model

```python
from sklearn.cluster import KMeans
model=KMeans(4,random_state=8)
y2=model.fit(x).labels_
plt.scatter(x[:,0],x[:,1],c=y2)
```

calculating the clusters' means

```python
import numpy as np
plt.scatter(x[:,0],x[:,1],c=y2,alpha=0.4)
Means=model.cluster_centers_
plt.scatter(Means[:,0],Means[:,1],s=100,c=np.unique(y2))
Means
```

here, we have 4 groups of subjects and the big dots, indicate the centers of each cluster and the more a point is closer (distance from the center and the cluster), it is classified as part of that group,for the points in between two groups must identified with only one group by using the euclidean distance methods to determine their distances with the respective dots. in the above data, the coloring of the points has helped us cluster the 3 points in between the purple group and red as belonging to the purple group.
